#Q.1
#Step 4. Discover what is the mean age per occupation

```python
mean_age = users.groupby(['occupation']).agg({'age':'mean'})
print("-----Average Occupation Age-----")
print(mean_age)
```

#Q.1
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least 

```python
male = users[users.gender > 'F']
male_sum = (male.groupby(['occupation'])['gender'].count()/users.groupby(['occupation'])['gender'].count()) #.sort_values(ascending = False)
print("-----Male Occupation Sorted in Descending Order-----")
print(male_sum)
```

#Q.1
#Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_age = users.groupby(['occupation']).agg({'age':'min'})
max_age = users.groupby(['occupation']).agg({'age':'max'})
print("-----Occupation by Minimum Age-----")
print(min_age)
print(' ')
print("-----Occupation by Maximum Age-----")
print(max_age)
```

#Q.1
#Step 7. For each combination of occupation and sex, calculate the mean age 

```python
mean_age = users.groupby(['occupation','gender']).agg({'age':'mean'})
print("-----Average Age by Occupation and Gender-----")
print(mean_age)
```

#Q.1
#Step 8. For each occupation present the percentage of women and men

```python
percent_m_f = (users.groupby(['occupation', 'gender'])['gender'].count().groupby(level = 0).transform(lambda x: x/x.sum()*100))
print("-----Percentage by Gender for Each Occupation-----")
print(percent_m_f)
```

#Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as pp
md = pd.read_csv('D:/Desktop/Problem Set 3/us-marriages-divorces-1867-2014.csv')
md.head(3)
```

```python
md['MPerCapita'] = (md['Marriages']/md['Population'])
md['DPerCapita'] = (md['Divorces']/md['Population'])
md.head(3)
```

```python
pp.plot(md.loc[:,'Year'], md.loc[:,'MPerCapita'], color = 'blue', label = 'Marriage per Capita')
pp.plot(md.loc[:,'Year'], md.loc[:,'DPerCapita'], color = 'red', label = 'Divorce per Capita')
pp.title('Marriages and Divorces Per Capita in the US') 
pp.xlabel('Years')
pp.ylabel('Marriage/Divorce Per capita')
pp.legend()
```
